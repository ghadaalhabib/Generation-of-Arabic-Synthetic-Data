```python
import os
from openpyxl import Workbook
from openpyxl.drawing.image import Image

# Create a new Excel workbook
workbook = Workbook()
sheet = workbook.active

# Define the directory containing the images
directory = './out'

# Get a list of the files in the directory
files = os.listdir(directory)
print(files)

# Define a lambda function to extract the sequence number prefix from the file name
sequence_number = lambda x: int(x.split(".")[0]) if x.split(".")[0].isdigit() else 0
# Sort the list of files based on the sequence number prefix
sorted_files = sorted(files, key=sequence_number)

# Print the sorted list of files
print(sorted_files)

# Loop through the files in the directory
for index, filename in enumerate(sorted_files):
    if filename.endswith('.jpg'):  # change this to the file extension of your images
        # Create a full file path and open the image
        print(filename)
        filepath = os.path.join(directory, filename)
        image = Image(filepath)

        # Add the image to the Excel sheet
        sheet.add_image(image, f'A{index+1}')  # add to column A, increment row for each image

# Save the Excel workbook
workbook.save('output.xlsx')
```
